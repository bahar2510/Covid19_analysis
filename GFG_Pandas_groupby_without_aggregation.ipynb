```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data=load_breast_cancer()
df=pd.DataFrame(cancer_data.data,columns=cancer_data.feature_names)
df['target']=pd.Series(cancer_data.target)
df.head()
```

```python
df['worst texture'].describe()
```

```python
df1=df[['worst texture','worst area','target']]
gr1=df1.groupby(df1['target']).mean()
gr1
```

```python
def meanofTargets(group1):
    wt=group1['worst texture'].agg('mean')
    wa=group1['worst area'].agg('mean')
    group1['Mean worst teture']=wt
    group1['Mean worst area']=wa
    return group1
```

```python
df2=df1.groupby('target').apply(meanofTargets)
df2
```

```python
df['mean area'].describe()
```

```python
def processed_area(df):
    label=['150-550','550-950','950-1350','1350-1750','1750-2150','2150-2500']
    cat_area=pd.cut(df['mean area'],bins=[150,550,950,1350,1750,2150,2500],labels=label)
    merge1=pd.concat([df,cat_area],axis=1)
    return merge1
```

```python
df_2=processed_area(df)
df_2.head()
```

```python
df_2.columns
```

```python
df_2.columns.values[31]='Cat_mean_area'
df_2.columns
```

```python
df_2['Cat_mean_area']
```

```python
gr=df_2.groupby(df_2['Cat_mean_area']).sum()
gr
```

```python
df_3=df_2[['mean area','Cat_mean_area','target']]
gr2=df_3.groupby(df_2['Cat_mean_area']).sum()
gr2
```

Now we will use a function to do the above sum aggregation without groupby.

```python
def totalTargets(group):
    g=group['target'].agg('sum')
    group['Total_targets']=g
    return group
```

```python
df_4=df_3.groupby(df_3['Cat_mean_area']).apply(totalTargets)
df_4
```
