```
covid.drop(['new_tests','new_tests_per_thousand'],axis=1,inplace=True)
```

```
covid
```

```
covid['tests_units']=covid['tests_units'].fillna('samples tested')
print(covid['tests_units'].isnull().sum(),
covid['tests_units'].notnull().sum())
```

```
covid['total_tests']=covid['total_tests'].fillna(covid['total_tests'].mean())
covid['total_tests_per_thousand']=covid['total_tests_per_thousand'].fillna(covid['total_tests_per_thousand'].mean())
covid['new_tests_smoothed']=covid['new_tests_smoothed'].fillna(covid['new_tests_smoothed'].mean())
covid['new_tests_smoothed_per_thousand']=covid['new_tests_smoothed_per_thousand'].fillna(covid['new_tests_smoothed_per_thousand'].mean())
covid['stringency_index']=covid['stringency_index'].fillna(covid['stringency_index'].mean())
sns.heatmap(covid.isnull())
```
